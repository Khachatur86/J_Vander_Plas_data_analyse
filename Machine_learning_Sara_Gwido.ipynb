```python
import pandas as pd
import numpy as np
import scipy
import matplotlib
# import mglearn
import sklearn
from sklearn.datasets import load_iris

iris_dataset = load_iris()

print('Ключи iris_dataset: \n {}'.format(iris_dataset.keys()))
print(iris_dataset['DESCR'])
print(f'Названия сортов:{iris_dataset["target_names"]}')
print(f"Названия признаков: \n {iris_dataset['feature_names']}")
# print(f"Массив data: {iris_dataset['data']}")
print(f"Тип массива data: {type(iris_dataset['data'])}")
print(f"Форма массива data: {iris_dataset['data'].shape}")
print(f"Тип массива target: {type(iris_dataset['target'])}")
print(f"Форма массива target: {iris_dataset['target'].shape}")

# Метрики эффективности: обучающий и тестовый наборы

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
  iris_dataset['data'], iris_dataset['target'], random_state=0
)


print(f"Массив X_train: {X_train}")
print(f"Массив y_train: {y_train}")
print(f"Массив X_test: {X_test}")
print(f"Массив y_test: {y_test}")
print(f"Форма массива X_train: {X_train.shape}")
print(f"Форма массива y_train: {y_train.shape}")
print(f"Форма массива X_test: {X_test.shape}")
print(f"Форма массива y_test: {y_test.shape}")

# Иллюистрация данных
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15),
                        marker='o', hist_kwds={'bins':20}, s=60, 
                        alpha=.8, cmap=mglearn.cm3)

```
